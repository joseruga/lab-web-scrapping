#### Display the trending Python repositories in GitHub

The steps to solve this problem is similar to the previous one except that you need to find out the repository names instead of developer names.

````python
# This is the url you will scrape in this exercise
url2 = 'https://github.com/trending/python?since=daily'
````

````python
html2 = requests.get(url2).content
soup_2 = BeautifulSoup(html2,'lxml')
titles = soup_2.select('h1[class="h3 lh-condensed"]')
````

````python
elements_list = [element.text for element in titles]
titles_list = [title[(title.index('/')+1):].strip() for title in elements_list]
print (titles_list)
````

````python
user_list = [user[:(user.index('/'))].strip() for user in elements_list]
nameandtitle = [title+' by '+user_list[i] for i,title in enumerate(titles_list)]
nameandtitle
````
